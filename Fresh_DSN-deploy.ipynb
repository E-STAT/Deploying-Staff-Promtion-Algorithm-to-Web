```python
from sklearn.preprocessing import LabelEncoder
#LabelEncoder?
```

```python
cat_features = (['Division', 'Qualification','Gender', 'Channel_of_Recruitment', 'Foreign_schooled', 'Marital_Status', 'Past_Disciplinary_Action', 
                'Previous_IntraDepartmental_Movement'])
```

```python
encoder = LabelEncoder()
encoded = redo_train[cat_features].apply(encoder.fit_transform)
encoded_test = redo_test[cat_features].apply(encoder.fit_transform)
```

```python
redo_train.drop(['Division', 'Qualification','Gender', 'Channel_of_Recruitment', 'Foreign_schooled', 'Marital_Status', 'Past_Disciplinary_Action', 
                'Previous_IntraDepartmental_Movement'], axis = 1, inplace = True)
redo_test.drop(['Division', 'Qualification','Gender', 'Channel_of_Recruitment', 'Foreign_schooled', 'Marital_Status', 'Past_Disciplinary_Action', 
                'Previous_IntraDepartmental_Movement'], axis = 1, inplace = True)
```

```python
redo_train= redo_train.join(encoded)
#print(redo_train.shape)

redo_test= redo_test.join(encoded_test)
#redo_test.shape
```

```python
No_of_previous_employers_ = pd.to_numeric(redo_train.No_of_previous_employers)
No_of_previous_employers__ = pd.to_numeric(redo_test.No_of_previous_employers)

redo_train.drop(['No_of_previous_employers'], axis = 1, inplace = True)
redo_test.drop(['No_of_previous_employers'], axis = 1, inplace = True)

redo_train = pd.concat([redo_train, No_of_previous_employers_], axis = 1)
redo_test = pd.concat([redo_test, No_of_previous_employers__], axis = 1)
```

```python
y = redo_train['Promoted_or_Not']
X = redo_train.drop(['Promoted_or_Not'], axis=1)
```

```python
import lightgbm as lgb
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc, accuracy_score, roc_auc_score, roc_curve, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)
X_train.columns
```

```python
estimator = lgb.LGBMClassifier(learning_rate = 0.15, metric = 'l1', 
                        n_estimators = 24, num_leaves = 38)


param_grid = {
    'n_estimators': [x for x in range(24,40,2)],
    'learning_rate': [0.10, 0.125, 0.15, 0.175, 0.2]}
gridsearch = GridSearchCV(estimator, param_grid)

gridsearch.fit(X_train, y_train,
        eval_set = [(X_test, y_test)],
        eval_metric = ['auc', 'binary_logloss'],
        early_stopping_rounds = 5,
              verbose = 1)
```

```python
print('Best parameters found by grid search are:', gridsearch.best_params_)
```

```python
gbm = lgb.LGBMClassifier(learning_rate = 0.125, metric = 'l1', 
                        n_estimators = 36)


gbm.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric=['auc', 'binary_logloss'],
early_stopping_rounds=5)
```
